```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/BostonHousing.csv')
df
```

```python
df.isna().any()
```

```python
df[df['rm'].isna()]
```

```python
df.dropna(subset=['rm'], inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# 2 Variables

```python
# Avarage rooms number and median price
sns.scatterplot(x='rm', y='medv', data=df)
plt.show()
```

```python
# Precentage of houses built before 1940 and median age
sns.lmplot(x='age', y='medv', data=df)
plt.show()
```

```python
#Precetage of lower status of the population and median price
sns.jointplot(x='lstat', y='medv', data=df, kind="hex")
plt.show()
```

# 3 variables

```python
# Avarage rooms number, median price, near charls lake
sns.scatterplot(x='rm', y='medv', hue='chas', data=df)
plt.show()
```

```python
# crime precentage, median price and low status precentage
sns.scatterplot(x='crim', y='medv',  hue='lstat', data=df)
plt.show()
```

```python
# low status precentage, median age and access to main roads
# We can explain low prices for houses near main roads by noise and polution
sns.scatterplot(x='lstat', y='medv', hue='rad', data=df)
plt.show()
```

```python

# number of rooms, median price, polution, presentage of low status
sns.scatterplot(x='rm', y='medv', hue='nox', size='lstat', data=df)
plt.show()
```
