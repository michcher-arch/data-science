```python
import pandas as pd

data = {
    'StudentID': [1, 2, 3, 4, 5],
    'Name': ['John', 'Emily', 'Robert', 'Linda', 'Michael'],
    'Age': [21, 22, 20, 23, 22],
    'Gender': ['M', 'F', 'M', 'F', 'M'],
    'GradeMath': [85, 90, 78, 92, 88],
    'GradeEnglish': [75, 92, 80, 88, 85],
    'GradeScience': [88, 85, 90, 84, 89]
}

df = pd.DataFrame(data)

df
```

```python
# write code here
df["Avarage"] = (df['GradeMath'] + df['GradeEnglish'] + df['GradeScience']) // 3
df
```
