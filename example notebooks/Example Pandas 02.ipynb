```python
import numpy as np
import pandas as pd
```

# Pandas DataFrame

```python
np.random.seed(42)
data1 = np.random.randint(-20,20,(5,4))
data1
```

```python
columns1 = ['A','B','C','D']
index1 = ['a','b','c','d','e']
df = pd.DataFrame(data=data1, index=index1, columns=columns1)
df
```

# Conditional selection

```python
df['B']>0
```

```python
df[ df['B']>0 ]
```

```python
(df['B']>0) & (df['C']<0)
```

```python
df[(df['B']>0) & (df['C']<0)]
```

# Missing Data

```python
df = pd.DataFrame({'A':[1,2,np.nan,4],
                  'B':[5,np.nan,np.nan,8],
                  'C':[9,10,11,12]})
df
```

```python
df['B'].dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(axis=0)
```

```python
df
```

```python
df['A'].fillna(df['A'].median())
```

```python
df.fillna(df.mean())
```

```python
df.fillna(value=0.0)
```
