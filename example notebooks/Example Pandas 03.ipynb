# Groupby

```python
df.groupby('color').median()
```

```python
df.groupby('color').mean().sort_index(ascending=False)
```

```python
df.groupby('color').mean().sort_values(by='carat',ascending=True)
```

```python
df.groupby('cut').max()[['depth']]
```

```python
df.groupby('cut').max()[['depth']].sort_index(ascending=False)
```

```python
df.groupby('cut').max()[['depth']].sort_values(by='depth',ascending=False)
```

```python
df.groupby('cut').describe()[['depth']]
```

```python
df.groupby('cut').describe()[['depth']].transpose()
```
